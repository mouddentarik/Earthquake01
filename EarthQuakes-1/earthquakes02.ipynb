```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
import plotly.express as px

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from datetime import datetime
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
Earthquakes3 = pd.read_csv("Earthquakes3.csv")
Earthquakes3.drop(Earthquakes3.columns[[0]], axis=1, inplace=True)

DataFrame_JupEarth = pd.read_csv("DataFrame_JupEarth.csv")
DataFrame_JupEarth.drop(DataFrame_JupEarth.columns[[0]], axis=1, inplace=True)

DataFrame_MarsEarth = pd.read_csv("DataFrame_MarsEarth.csv")
DataFrame_MarsEarth.drop(DataFrame_MarsEarth.columns[[0]], axis=1, inplace=True)

DataFrame_MoonEarth = pd.read_csv("DataFrame_MoonEarth.csv")
DataFrame_MoonEarth.drop(DataFrame_MoonEarth.columns[[0]], axis=1, inplace=True)

DataFrame_NeptuneEarth = pd.read_csv("DataFrame_NeptuneEarth.csv")
DataFrame_NeptuneEarth.drop(DataFrame_NeptuneEarth.columns[[0]], axis=1, inplace=True)

DataFrame_SaturneEarth = pd.read_csv("DataFrame_SaturneEarth.csv")
DataFrame_SaturneEarth.drop(DataFrame_SaturneEarth.columns[[0]], axis=1, inplace=True)

DataFrame_SunEarth = pd.read_csv("DataFrame_SunEarth.csv")
DataFrame_SunEarth.drop(DataFrame_SunEarth.columns[[0]], axis=1, inplace=True)

DataFrame_UranEarth = pd.read_csv("DataFrame_UranEarth.csv")
DataFrame_UranEarth.drop(DataFrame_UranEarth.columns[[0]], axis=1, inplace=True)


DataFrame_venusEarth = pd.read_csv("DataFrame_venusEarth.csv")
DataFrame_venusEarth.drop(DataFrame_venusEarth.columns[[0]], axis=1, inplace=True)

DataFrame_MercureEarth = pd.read_csv("DataFrame_MercureEarth.csv")
DataFrame_MercureEarth.drop(DataFrame_MercureEarth.columns[[0]], axis=1, inplace=True)
```
